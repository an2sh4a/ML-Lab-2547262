```python
print(city.isnull().sum())
print()
print(crop.isnull().sum())
```

```python
city_missing = (city.isnull().sum() / len(city)) * 100
crop_missing = (crop.isnull().sum() / len(crop)) * 100

print("City Dataset Missing Percentage")
print(city_missing)
print("\nCrop Dataset Missing Percentage")
print(crop_missing)
```

```python
missing_cols = city_missing[
    city_missing > 50
].index
print("Columns with more than 50% missing values:")
print(missing_cols)
```

```python
city = city.drop(columns=missing_cols)
print(city.columns)
```

```python
num_cols = city.select_dtypes(include=['int64','float64']).columns
cat_cols = city.select_dtypes(include=['object']).columns
print("Numerical Columns - ")
print(num_cols)
print("\nCategorical Columns - ")
print(cat_cols)
```

```python
for col in num_cols:
    if city[col].isnull().sum() > 0:
        city[col] = city[col].fillna(
            city[col].median()
        )

for col in cat_cols:
    if city[col].isnull().sum() > 0:
        city[col] = city[col].fillna(
            city[col].mode()[0]
            
        )

print(city.isnull().sum())
```

```python
crop['Production'] = crop['Production'].fillna(
    crop['Production'].median()
)
print(crop.isnull().sum())
```

### Missing Value Strategy
Approach I used to handle misisng values:

Step 1:
Calculated the percentage of missing values for every column

Step 2:
Columns with more than 50% missing values were dropped because this could introduce significant bias and reduce reliability.

Step 3:
For the remaining numerical columns, missing values were replaced using the median. Not mean because pollution and agricultural measurements contain extreme values and skewed distributions.

Step 4:
For categorical columns, missing values were replaced using the mode .

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Are most cities moderately polluted or are only a few cities heavily polluted?

plt.figure(figsize=(10,5))
sns.histplot(city['AQI'], bins=30, kde=True, color='brown')
plt.title("Distribution of AQI Values")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

A histogram because it shows how AQI values are distributed and helps identify the range in which most observations are concentrated.

### Observation

- Most AQI observations are concentrated between 50 and 250, indicating that the majority of cities experience moderate to poor air quality levels.
- The distribution is positively skewed (right-skewed), with a long tail extending toward very high AQI values.
- Very few observations have AQI values above 500, suggesting that extreme pollution events are relatively uncommon.

### Conclusion

The histogram shows that air pollution is not evenly distributed across observations. Most AQI readings fall within moderate ranges, while a smaller number of extremely high AQI values create a long right tail in the distribution.

```python
# Is the average AQI being unfairly increased by a few extreme cities?
plt.figure(figsize=(10,4))
sns.boxplot(x=city['AQI'], color='purple')
plt.title("Boxplot of AQI Values")
plt.xlabel("AQI")
plt.show()
```

A boxplot was chosen because it clearly identifies outliers and helps determine whether extreme AQI values may influence summary statistics such as the mean.

### Observation

- The boxplot reveals a large number of outliers beyond the upper whisker.
- Several AQI values exceed 1000, with the highest values crossing 2000.
- The spread of the data is substantial, indicating considerable variation in air quality across cities and time periods.
- The mean AQI (158.78) is significantly higher than the median AQI (118), confirming the influence of extreme values.

### Conclusion

The boxplot confirms the presence of numerous extreme AQI values. These outliers increase the average AQI and make the mean less representative of typical air quality conditions. The median provides a more reliable measure of the central tendency of AQI in this dataset.

```python
Q1 = city['AQI'].quantile(0.25)

Q3 = city['AQI'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
outliers = city[
    (city['AQI'] < lower_bound) |
    (city['AQI'] > upper_bound)
]
print("Number of Outliers:", len(outliers))
```

```python
outlier_percentage = (
    len(outliers) / len(city)
) * 100

print("Outlier Percentage:", round(outlier_percentage,2), "%")
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    x=city['AQI'],
    color='orange'
)
plt.title("AQI Before Outlier Treatment")
plt.show()
```

### Observation

- The IQR method identified 3,192 outlier observations in the AQI dataset.
- These outliers account for approximately 10.81% of all records.
- Most outliers occur on the upper side of the distribution, indicating unusually high pollution levels.
- The boxplot shows a large number of extreme AQI values extending far beyond the upper whisker.
- The presence of these extreme observations explains the strong right-skew observed in Task 4.

```python
city['AQI_Clean'] = city['AQI'].clip(
    lower=lower_bound,
    upper=upper_bound
)
```

We can't delete outlier rows as
- We will lose genuine data.
- AQI values above 500 are real pollution events.

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    x=city['AQI_Clean'],
    color='maroon'
)
plt.title("AQI After Outlier Treatment")
plt.xlabel("AQI")
plt.show()
```
