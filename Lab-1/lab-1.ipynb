```python
print(city.isnull().sum())
print()
print(crop.isnull().sum())
```

```python
city_missing = (city.isnull().sum() / len(city)) * 100
crop_missing = (crop.isnull().sum() / len(crop)) * 100

print("City Dataset Missing Percentage")
print(city_missing)
print("\nCrop Dataset Missing Percentage")
print(crop_missing)
```

```python
missing_cols = city_missing[
    city_missing > 50
].index
print("Columns with more than 50% missing values:")
print(missing_cols)
```

```python
city = city.drop(columns=missing_cols)
print(city.columns)
```

```python
num_cols = city.select_dtypes(include=['int64','float64']).columns
cat_cols = city.select_dtypes(include=['object']).columns
print("Numerical Columns - ")
print(num_cols)
print("\nCategorical Columns - ")
print(cat_cols)
```

```python
for col in num_cols:
    if city[col].isnull().sum() > 0:
        city[col] = city[col].fillna(
            city[col].median()
        )

for col in cat_cols:
    if city[col].isnull().sum() > 0:
        city[col] = city[col].fillna(
            city[col].mode()[0]
            
        )

print(city.isnull().sum())
```

```python
crop['Production'] = crop['Production'].fillna(
    crop['Production'].median()
)
print(crop.isnull().sum())
```

### Missing Value Strategy
Approach I used to handle misisng values:

Step 1:
Calculated the percentage of missing values for every column

Step 2:
Columns with more than 50% missing values were dropped because this could introduce significant bias and reduce reliability.

Step 3:
For the remaining numerical columns, missing values were replaced using the median. Not mean because pollution and agricultural measurements contain extreme values and skewed distributions.

Step 4:
For categorical columns, missing values were replaced using the mode .

```python
Q1 = city['AQI'].quantile(0.25)

Q3 = city['AQI'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
outliers = city[
    (city['AQI'] < lower_bound) |
    (city['AQI'] > upper_bound)
]
print("Number of Outliers:", len(outliers))
```

```python
outlier_percentage = (
    len(outliers) / len(city)
) * 100

print("Outlier Percentage:", round(outlier_percentage,2), "%")
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    x=city['AQI'],
    color='orange'
)
plt.title("AQI Before Outlier Treatment")
plt.show()
```

### Observation

- The IQR method identified 3,192 outlier observations in the AQI dataset.
- These outliers account for approximately 10.81% of all records.
- Most outliers occur on the upper side of the distribution, indicating unusually high pollution levels.
- The boxplot shows a large number of extreme AQI values extending far beyond the upper whisker.
- The presence of these extreme observations explains the strong right-skew observed in Task 4.

```python
city['AQI_Clean'] = city['AQI'].clip(
    lower=lower_bound,
    upper=upper_bound
)
```

We can't delete outlier rows as
- We will lose genuine data.
- AQI values above 500 are real pollution events.

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    x=city['AQI_Clean'],
    color='lightgreen'
    
)
plt.title("AQI After Outlier Treatment")
plt.xlabel("AQI")
plt.show()
```

### Observation

- The boxplot after treatment shows a much more compact AQI distribution.
- Extreme AQI values beyond the upper whisker are no longer visible.
- The spread of the data has reduced considerably compared to the original distribution.
- All observations were retained, but extreme values were capped at the IQR limits.
- Recalculation confirmed that the percentage of outliers decreased from 10.81% to 0%.

```python
print(city[['AQI','AQI_Clean']].describe())
```

```python
new_outliers = city[
    (city['AQI_Clean'] < lower_bound) |
    (city['AQI_Clean'] > upper_bound)
]

print("Remaining Outliers:", len(new_outliers))
```

The number of outliers reduced from 3,192 (10.81%) before treatment to 0 (0%) after treatment, confirming the effectiveness of the IQR-based capping approach.

Is India's air getting better or worse over time?

Why use a Line Chart?
- Years are ordered chronologically.
- To observe trends over time.
- Line charts are the best choice for time-series analysis.

```python
#coverting date to datetime
print(city['Date'].dtype)
```

```python
city['Date'] = pd.to_datetime(city['Date'])
print(city['Date'].dtype)
```

```python
#extracting year from date
city['Year'] = city['Date'].dt.year
print(city[['Date','Year']].head())
```

```python
#Average AQI per Year
yearly_aqi = city.groupby('Year')['AQI_Clean'].mean()
print(yearly_aqi)
```

```python
most_polluted_year = yearly_aqi.idxmax()
least_polluted_year = yearly_aqi.idxmin()

print("Most Polluted Year:", most_polluted_year)
print("AQI:", yearly_aqi.max())

print("\nLeast Polluted Year:", least_polluted_year)
print("AQI:", yearly_aqi.min())
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    x=yearly_aqi.index,
    y=yearly_aqi.values,
    marker='o',
    color='magenta'
)

plt.scatter(
    most_polluted_year,
    yearly_aqi.max(),
    s=150,
    color='red',
    label='Most Polluted Year'
)

plt.scatter(
    least_polluted_year,
    yearly_aqi.min(),
    s=150,
    color='green',
    label='Least Polluted Year'
)

plt.title("Average AQI Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average AQI")
plt.legend()
plt.show()
```

### Observation

- The average AQI shows an overall decreasing trend from 2015 to 2020.
- The highest average AQI was recorded in 2015 (165.87), indicating poorer air quality during that year.
- AQI decreased slightly from 2015 to 2017, followed by a small increase in 2018.
- After 2018, the AQI declined significantly, reaching 140.59 in 2019.
- The lowest average AQI was observed in 2020 (110.87).
- The line graph indicates a substantial improvement in air quality over the study period.

### Conclusion

Air quality have improved over time. The average AQI decreased from 165.87 in 2015, the most polluted year in the dataset, to 110.87 in 2020, the least polluted year. 
Although there was a small increase in pollution during 2018, the overall trend shows a steady decline in AQI levels. This suggests that air pollution became less severe. While the data alone cannot prove which specific policy caused the improvement, the reduction in average AQI after 2018 is consistent with the possibility that pollution control efforts may have contributed to better air quality.


```python
aqi_improvement = (
    (yearly_aqi[2015] - yearly_aqi[2020])
    / yearly_aqi[2015]
) * 100

print(round(aqi_improvement, 2))
```

The average AQI decreased by approximately 33.16% between 2015 and 2020, indicating a improvement in air quality.

### Transformation Required Before Merge

The AQI dataset contains records from 2015–2020, while the crop dataset contains records from 1997–2015. Therefore, the two datasets overlap only in the year 2015.

Additionally, the AQI dataset is recorded at the city-day level whereas the crop dataset is recorded at the state-year level. To make them compatible, AQI data must be aggregated to the state level and the crop data must be filtered to the common year (2015) before merging.

The merge will therefore be performed using the State column for the year 2015.

```python
#State-wise AQI Dataset
aqi_state = city.groupby(
    'State'
)['AQI_Clean'].mean().reset_index()
aqi_state.head(10)
```

```python
#State-wise Crop Dataset
crop_state = crop.groupby(
    'State_Name'
)['Production'].sum().reset_index()
crop_state.head(10)
```

```python
crop_state.rename(
    columns={'State_Name':'State'},
    inplace=True
)
```

```python
print(aqi_state.shape)

print(crop_state.shape)
```

```python
merged_df = pd.merge(
    aqi_state,
    crop_state,
    on='State',
    how='inner'
)
print(merged_df.shape)
merged_df.head(10)
```

```python
#Correlation Analysis
corr_matrix = merged_df.corr(
    numeric_only=True
)
print(corr_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='Reds'
)
plt.title(
    'Correlation Between AQI and Crop Production'
)
plt.show()
```

### Observations

- The AQI and crop datasets were successfully transformed to the state level and merged using the State column as the common key.
- The merged dataset contained 20 states with information on both average AQI and crop production.
- The correlation matrix shows a negative correlation (-0.252) between AQI and crop production.
- This suggests that states with higher pollution levels generally tend to have lower crop production.
- However, the correlation is weak, indicating that air pollution alone does not fully explain differences in agricultural output across states.

### Conclusion

After transforming both datasets to the state level and merging them using the State column, a weak negative relationship (-0.252) was observed between AQI and crop production. This suggests that higher pollution may be associated with lower agricultural output, but the relationship is not strong enough to conclude that pollution is the primary cause. Additional factors such as rainfall, irrigation, soil quality, crop selection, and economic conditions likely influence agricultural production and are not captured in the available datasets.
